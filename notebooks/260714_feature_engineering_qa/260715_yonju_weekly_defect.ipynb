```python
import polars as pl

input_path1 = "/data01/Documents/0_hw/notebooks/data_pipeline_archived/data_pipeline_v5/data/TRACEABILITY_260201_260331.parquet"
input_path2 = "/data01/Documents/0_hw/notebooks/datamart_pipeline_v3/data/flattened/Y_NFF_Grade.parquet"
output_path = "/data01/Documents/98_model/notebooks/260714_feature_engineering_qa/weekly_defect_n32s.parquet"
join_key = "07_Before Degas_Cell ID"
lot_id_col = "06_Assembly_Lot ID"

lf1 = pl.scan_parquet(input_path1)
lf2 = pl.scan_parquet(input_path2)

lf = lf1.join(lf2, on=join_key, how="left")

schema = lf.collect_schema()

numeric_types = [pl.Int8, pl.Int16, pl.Int32, pl.Int64, pl.UInt8, pl.UInt16, pl.UInt32, pl.UInt64, pl.Float32, pl.Float64]
y_upper_cols = sorted([c for c, dtype in schema.items() if c.startswith("Y_") and dtype in numeric_types])
y_lower_cols = sorted([c for c, dtype in schema.items() if c.startswith("y_") and dtype in numeric_types])
target_cols = y_upper_cols + y_lower_cols

result = (
    lf
    .explode(lot_id_col)
    .with_columns([
        pl.col(lot_id_col).str.slice(3, 1).alias("_year_char"),
        pl.col(lot_id_col).str.slice(4, 1).str.to_lowercase().alias("_month_char"),
        pl.col(lot_id_col).str.slice(5, 2).alias("_day_str"),
    ])
    .with_columns([
        pl.when(pl.col("_year_char") == "E").then(2025)
          .when(pl.col("_year_char") == "F").then(2026)
          .when(pl.col("_year_char") == "G").then(2027)
          .otherwise(None)
          .alias("_year"),
        pl.when(pl.col("_month_char") == "a").then(1)
          .when(pl.col("_month_char") == "b").then(2)
          .when(pl.col("_month_char") == "c").then(3)
          .when(pl.col("_month_char") == "d").then(4)
          .when(pl.col("_month_char") == "e").then(5)
          .when(pl.col("_month_char") == "f").then(6)
          .when(pl.col("_month_char") == "g").then(7)
          .when(pl.col("_month_char") == "h").then(8)
          .when(pl.col("_month_char") == "i").then(9)
          .when(pl.col("_month_char") == "j").then(10)
          .when(pl.col("_month_char") == "k").then(11)
          .when(pl.col("_month_char") == "l").then(12)
          .otherwise(None)
          .alias("_month"),
        pl.col("_day_str").str.extract(r"^(\d{2})$", 1).cast(pl.Int32).alias("_day"),
    ])
    .filter(pl.col("_year").is_not_null() & pl.col("_month").is_not_null() & pl.col("_day").is_not_null())
    .with_columns(pl.date("_year", "_month", "_day").alias("_date"))
    .with_columns([
        pl.col("_date").dt.iso_year().alias("year"),
        pl.col("_date").dt.week().alias("week"),
        (pl.col("_date") - pl.duration(days=pl.col("_date").dt.weekday())).alias("_week_start"),
    ])
    .drop([lot_id_col, "_year_char", "_month_char", "_day_str", "_year", "_month", "_day", "_date"])
    .group_by(["year", "week"])
    .agg([
        pl.col("_week_start").first().alias("_week_start"),
        *[pl.col(c).mean().alias(c) for c in target_cols]
    ])
    .sort(["year", "week"])
    .with_columns(
        (pl.col("_week_start") + pl.duration(days=6)).alias("_week_end")
    )
    .with_columns(
        (pl.col("_week_start").dt.strftime("%Y.%m.%d") + pl.lit("-") + pl.col("_week_end").dt.strftime("%Y.%m.%d")).alias("date")
    )
    .drop(["_week_start", "_week_end", "year"])
    .select(["date", "week"] + target_cols)
)

result.sink_parquet(output_path)
print(f"저장 완료: {output_path}")
```
